```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
datos = pd.read_csv("Original Data/all_seasons.csv")
datos
```

```python
datos_españa = datos[datos['country'] == 'Spain'] 
datos_españa
```

```python
datos_españa.columns
```

```python
por_temporada = datos_españa.groupby('season').size()

fig, ax = plt.subplots()
por_temporada.plot(kind="line", marker="o", ax=ax, color="red")


plt.ylabel("Total players")
plt.xlabel("Season")
plt.title('Spanish players by season NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.1, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```

```python
por_gp = datos_españa.groupby('player_name').size()
top_10_gp = por_gp.nlargest(10)

fig, ax = plt.subplots()
top_10_gp.plot(kind="pie", ax=ax)


plt.title('Spanish players by number of seasons (Top 10) NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.5, 0.5),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```

```python
por_draft_number = datos_españa.groupby('draft_year').size()

fig, ax = plt.subplots()
por_draft_number.plot(kind="bar", ax=ax, color="red")


plt.xlabel("Draft year")
plt.ylabel("Total players")
plt.title('Spanish players by draft year NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.9, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```
