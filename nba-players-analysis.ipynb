```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
datos = pd.read_csv("Original Data/all_seasons.csv")
datos
```

```python
datos.columns
```

```python
datos.describe()
```

```python
undrafted_datos = datos[datos["draft_year"] == 'Undrafted']
undrafted_datos
```

```python
undrafted_datos.describe()
```

```python
undrafted_datos = undrafted_datos[undrafted_datos['gp'] >= undrafted_datos['gp'].quantile(0.33)]
undrafted_datos
```

```python
jugadores_temporada = undrafted_datos.groupby('season').size()

fig, ax = plt.subplots()
jugadores_temporada.plot(kind="line", marker="o", color="orange", ax=ax)



plt.xlabel('Season')
plt.ylabel('Total players')
plt.title('Undrafted players with at least 33% games played by season NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.1, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```
