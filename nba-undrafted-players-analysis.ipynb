```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
datos = pd.read_csv("Original Data/all_seasons.csv")
datos
```

```python
datos.columns
```

```python
datos.describe()
```

```python
undrafted_datos = datos[datos["draft_year"] == 'Undrafted']
undrafted_datos
```

```python
undrafted_datos.describe()
```

```python
undrafted_datos = undrafted_datos[undrafted_datos['gp'] >= undrafted_datos['gp'].quantile(0.33)]
undrafted_datos
```

```python
jugadores_temporada = undrafted_datos.groupby('season').size()

fig, ax = plt.subplots()
jugadores_temporada.plot(kind="line", marker="o", color="orange", ax=ax)



plt.xlabel('Season')
plt.ylabel('Total players')
plt.title('Undrafted players with at least 33% games played by season NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.1, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```

```python
jugadores_edad = undrafted_datos.groupby('age').size()

fig, ax = plt.subplots()
jugadores_edad.plot(kind="bar", color="orange", ax=ax)



plt.xlabel('Age')
plt.ylabel('Total players')
plt.title('Undrafted players by age NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.9, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```

```python
jugadores_college = undrafted_datos.groupby('college').size()
top_10_colleges = jugadores_college.nlargest(10)

fig, ax = plt.subplots()
top_10_colleges.plot(kind="pie", ax=ax)


plt.title('Undrafted players by college (Top 10) NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (0.5, 0.5),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```

```python
jugadores_pais = undrafted_datos.groupby('country').size()
top_5_countries = jugadores_pais.nlargest(5)

fig, ax = plt.subplots()
top_5_countries.plot(kind="pie", ax=ax)



plt.title('Undrafted players by country (Top 5) NBA')

image_path = 'images/nba-logo-transparent.png'  # Reemplaza esto con la ruta a tu imagen
image = plt.imread(image_path)

imagebox = OffsetImage(image, zoom=0.015)  # Ajusta el 'zoom' para cambiar el tamaño de la imagen
ab = AnnotationBbox(imagebox, (1, 0.8),  # Cambia las coordenadas para posicionar la imagen
                    xycoords='axes fraction',
                    frameon=False)
ax.add_artist(ab)

plt.show()
```
